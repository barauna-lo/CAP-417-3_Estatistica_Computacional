# **1.3 Evolution Series of Means and SD for increasing subsets**

It seems that there is a theoretical limit near 1% of the population, after which both the mean and the standard deviation seem to respond very well to random sets, stabilizing around 10 sets.
This would imply that up to 10% of the population could be sampled in this pattern, i.e., as in the test the population was 10,000: 1% tested was 100 per subset times 10 sets randomly drawn from the population, approximately 10,000 values tested.

The range does not matter for the results, always converging in good estimates near 1% and 10 subsets 

For this example, a *tibble* (like a datframe, but for Tidyverse) was created with the evolution of the means and standard deviation with each new group added; thus, when presenting n = 5, the mean was in 5 subsets, n = 7 mean in 7 subsets, and so on.

**Some parameters:**

p = 10000                        <- total size of population (range = [ -200, 200 ])

s = [10, 20, 50, 100, 200, 250]  <- size of each subset

n = [1: 30]                      <- Number os subsets

Line graphs make it easy to see when there is convergence to the mean values of the full set (dotted black line).

```r
p <- 10000
set.seed(10); u.completo <- floor(runif(p, -200, 200))

wide <- options(repr.plot.width=24, repr.plot.height=8)
#ggplot () +
#  geom_line(aes(x=1:p, y=u.completo))
#options (wide)

u.mean <- mean(u.completo)
u.sd <- sd(u.completo)



u.sub <- tibble( values = numeric (),
                 n_total= numeric (),
                 n      = numeric (),
                 s      = numeric (),
                 mean   = numeric (),
                 sd     = numeric () 
)
for (s in c(p*0.1/100, p*0.2/100, p*0.5/100, p*1/100, p*2/100, p*2.5/100)){
  n <- 50
  m <- 0; a<- 0
  for (i in 1:30){
    set.seed(i); sub <- sample (u.completo, s)
    m <- m + mean (sub)
    a <- a + sd (sub)
#    print (c(s, i, sd(sub), a, a/i))
    u.sub <- u.sub %>% add_row( values  = sub, 
                                n_total = n, 
                                n       = i, 
                                s       = s,
                                mean    = m/i, 
                                sd      = a/i)
  }
}

p1 <- ggplot () +
  geom_line (data = u.sub, aes (x = as.numeric(n), y = mean, color = as.factor(s))) +
  geom_hline (aes(yintercept = mean(u.completo)), linetype = 'dotted') +
#  ylim (-150, 150) +
  theme(text = element_text(size=20)) +
  xlab("Random Subsets (count)") + ylab("Mean") + labs(color = "Subset Size") +
  ggtitle (paste("Mean of Subset Sizes X Subset Counts")) 


p2 <- ggplot () +
  geom_line (data = u.sub, aes (x = as.numeric(n), y = sd, color = as.factor(s))) +
  geom_hline (aes(yintercept = sd(u.completo)), linetype = 'dotted') +
#  ylim (0, 200) +
  theme(text = element_text(size=20)) +
  xlab("Random Subsets (count)") + ylab("Standard Deviation") + labs(color = "Subset Size") +
  ggtitle (paste("Standard Deviation of Subset Sizes X Subset Counts")) 


plot_grid (p1, p2, nrow=2)
```

# **2. Confidence Interval**
Normal distribution, with mean 186 and standard deviation 12, find the 95% confidence interval.

For this exercise, there are automatic alternatives for finding the values in x that limit the area of confidence, but google.colab was unable to use them, forcing a new approach for the solution.

A function describing the *normal function* has been implemented, and one can plot the function with the standard deviation and mean indicated in the exercise.

The initial value was + or - 3 times the standard deviation from the mean, x discretized to 0.05 and then applied to the density function *fdensity*.

To calculate the area under the curve, the MESS package was used, and the *auc* function with the *splines* method. I could have implemented the simpson method or the trapezoidal rule, but for brevity the *auc* function was used.

**Total AUC of *fdensity* = 0.9973**

The total probability should be = 1, i.e. the probability of any event occurring.

If the interval is 95%, it must be centered on the mean, then 2.5% to the left and the same to the right, so the area under the curve of interest must have a value of approximately 0.025.

To find the value that would correspond to the 2.5% of the curve's area, a loop was implemented where the area over the curve was scanned until it approached the value of 0.025. This loop started from the beginning of the discretization with increments of 0.05 and stop conditions of *area >= 0.0245 & area <= 0.0255*.

**x_ini = 162.65**

Com o valor de x correspondente aos 2.5% (162.65) da área e o x deste valor encontrado, foi calculado o *Z-score*, que representa quantos desvios padrão a partira da média está o intervalo de confiança; ou seja, + ou - Z-score aplicado na média define nosso intervalo de interesse.

**Z-score = 1.95**

**x_end = 209.35**

Again the *auc* function was used to check if the area around the curve was as expected, and received a response of approximately 95%.



```r
install.packages ('MESS')
require (ggplot2)
require(MESS)  # for AUC - spline

```

```r

fdensity <- function (x, mean, sd) {
  a <- 1/(sd*sqrt(2*pi))
  b <- -(x-mean)^2/(2*sd^2)
  return (a * exp (b))
}

dist.mean <- 186
dist.sd <- 12

dist.ini <- dist.mean - 3 * dist.sd
dist.end <- dist.mean + 3 * dist.sd

x <- seq (dist.ini, dist.end, 0.05)
y <- fdensity(x, dist.mean, dist.sd)
dist.area <- auc(x,y, type = 'spline')

print (c('Total AUC', dist.area))

for (i in seq(dist.ini + dist.sd/2, dist.ini + 2*dist.sd, 0.05)) {
  x1 <- x[x>=dist.ini & x<=i]
  y1 <- y[x>=dist.ini & x<=i]

  a2 <- auc(x1,y1, type = 'spline')
  parea <- a2/dist.area
#  print (c('AUC OK', i, a2))
  if (parea >= 0.0245 & parea <= 0.0255) {
    break
  }
}

dist.p025 <- i
dist.zscore <- -(dist.p025 - dist.mean)/dist.sd

# zscore <- (x - mean)/sd
# qnorm will return the value at 0.975 (97.5%)
# --2.5%|--------95%--------|2.5%--
# --------------97.5%-------|2.5%--

print (c("2.5% AUC", dist.p025, a2/dist.area))
print (c("Z-score", dist.zscore))

dist.p0975 <- dist.mean + dist.zscore * dist.sd
print (c(dist.p025, dist.p0975))

x1 <- x[x>=dist.p025 & x<=dist.p0975]
y1 <- y[x>=dist.p025 & x<=dist.p0975]
area <- auc(x1,y1, type = 'spline')

print (c("95% AUC", area))

saved <- options(repr.plot.width=10, repr.plot.height=10)
ggplot () +
  geom_line (aes(x=x, y=y)) +
  geom_area (aes(x=x1, y=y1), fill='lightblue', alpha =1) +
  geom_vline(aes(xintercept = seq(dist.mean-3*dist.sd, dist.mean+3*dist.sd, dist.sd)), alpha=0.2, linetype = 'dotted') +
  geom_vline(aes(xintercept = c(dist.p025, dist.p0975)), linetype = 'dotted', color = 'blue', size=1) +
  geom_hline(aes(yintercept = 0)) + 
  theme_bw() +
  theme(text = element_text(size=20), plot.title = element_text(hjust = 0.5)) +
  xlab("x") + ylab("y") + labs(color = "a, b") +
  ggtitle (paste("Normal Distribution, Mean = 186, Standard Deviation = 12\n ", area*100, "% (Blue)")) +
  geom_label (aes(x = dist.p025, y= -0.001, label = dist.p025), size = 8) +
  geom_label (aes(x = dist.p0975, y= -0.001, label = dist.p0975), size = 8) +
  geom_label (aes(x = dist.mean, y= 0.02, label = "95%"), size = 8)
options (saved)
```
