# Descriptive  Statistic

Now we generate a list of 5 and 100 integers and positives, which will be calculated the min, average, mode, median, max, range, sum, variance and standart deviation for each list.

First, we can define the calculations for each of the quoted values.

```python
def Min(lst):
  return np.min(lst)

def Average(lst):
  return sum(lst)/len(lst)

def Mode(lst):
  return max(set(lst), key = lst.count)   

def Median(lst):
  return np.median(lst)  

def Max(lst):
  return max(lst)    

def Range(lst):
  return len(lst)  

def Sum(lst):
  return sum(lst)   

def Variance(lst):
  return np.var(lst)  

def Standart_deviation(lst):
  return np.std(lst)   

def Percentile(lst, score):
  return np.percentile(lst, score)  
```

To explain how the percentile works, it's a score that evaluates the percentage of the data encompassed by that amount. For example, with $score = 95\%$, all $95\%$ of the points are closer to the median will be selected. The percentile takes the values closer to the median. Therefore, with a $score = 50$, the percentile is equal to the median.

## The list of 5 elements:

```python
rnd.seed(10)

lst = []
score = 50 #Range [0,99]

lst = [rnd.randint(1,10) for i in range(5)]

# List
print("List:", lst)

# Min
print("Min:", Min(lst))

# Average
print("Average:", Average(lst))

# Mode
print("Mode:", Mode(lst))

# Median
print("Median:", Median(lst))

# Max
print("Max:", Max(lst))

# Range
print("Range:", Range(lst))

# Sum
print("Sum:", Sum(lst))

# Variance
print("Variance:", Variance(lst))

# Standart Deviation
print("Standart Deviation:", Standart_deviation(lst))

# Percentile
print("Percentile:", Percentile(lst, score))
```

## Average as weighted sum

Thinking of the mean as the weighted sum of the values ‚Äã‚Äãin the list. One idea I had was to use as a weight for the sum, the percentage of each element related to the maximum value of the list. For example, in a list of 5 elements ($e=5$), with values ‚Äã‚Äãfrom 1 to 10, we randomly generate a list $[1,7,8,10,10]$.
Since the weighted sum ($S_p$) is the sum of each element in the list multiplied by its weight, we would have the following: $S_p = 1*0.1+7*0.7+8*0.8+10*1+10*1$.
So the average of the weighted sum ($W_p$) would be $S_p$ divided by the number of elements in the sequence, $W_p = S_p/e$.

## Bar Graphs (Horizontal and Vertical)

To generate a bar graph, whether vertical or horizontal, we need data that meets the requirements for it. For this, from the same list of 100 random numbers used earlier, we will evaluate how many numbers contained in the list are odd and how many are even.

```python
even = 0
odd = 0

for i in range(0,100):
  if(lst[i]%2 != 0):
    odd+=1
  else:
    even+=1  

duo = [even, odd]

bars = ("Even", "Odd")
y_pos = np.arange(len(bars))
```

With the numbers properly cataloged, we can generate a graph with two bars. One to represent the odd numbers and the other the even ones, where the size of each one represents the percentage of numbers that belong to each set, respectively.

```python
plt.figure()
# Create bars
plt.bar(y_pos, duo, color = ["firebrick", "navy"])

# Add title and axis names
plt.title("Bar Graph Vertical")
plt.ylabel("Percent")

# Create names on the x-axis
plt.xticks(y_pos, bars)

# Show graphic
plt.show()
```

Now, for the horizontal bars.

```python
plt.figure()
# Create bars
plt.barh(y_pos, duo, color = ["firebrick", "navy"])

# Add title and axis names
plt.title("Bar Graph Horizontal")
plt.xlabel("Percent")

# Create names on the x-axis
plt.yticks(y_pos, bars)

# Show graphic
plt.show()
```

# Moving Average

In this section, the moving average of a time series was calculated

Initially, we generate a random time series.

```python
#Time Series

rnd.seed(111)

lst1 = []
lst1 = [rnd.randint(1,100) for i in range(100)]

t = []
t = [i for i in range(100)]
```

Next, we will calculate the moving average. For this, will be define a fixed-size window and we will divide the list of generated random numbers into several sub-lists. The selection of sliding window depends on the amount of smoothing.

 After calculate the mean of this sub-list and we will walk with that window to the next data interval. Finally, all generated averages will be transformed into a new list and printed together in a graph for visualization.

```python
def MovingAverage(lst1, window):
  lst2 = []
  for i in range(len(lst1)-window+1):
    lst2.append(Sum(lst1[i:i+window])/window)  
  return lst2  

lst2 = MovingAverage(lst1, 3)

plt.figure()
plt.plot(range(len(lst1)), lst1, 'o')
plt.plot(np.arange(len(lst2))+1, lst2)
plt.title("Moving Average")
plt.xlabel("Sample Number")
plt.ylabel("Values")
plt.show()
```
