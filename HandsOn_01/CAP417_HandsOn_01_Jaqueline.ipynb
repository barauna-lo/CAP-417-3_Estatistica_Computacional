**Description of Libraries:**

The **numpy.random** library [1] was used to generate the random numbers.

```python
from numpy.random import Generator, PCG64
from matplotlib import pyplot as plt
```

**Description of code:**

The code below uses **seed_1** as **seed** to be used in two generators: **generator_1** and **generator_2**.

Both generators are of type Permuted Congruential Generator (64-bit, PCG64) [3], and have a period of $2^{128}$.

The "numpy.random.Generator.integers" method [4] returns random integers from a **"discrete uniform" distribution**.

Thus, the generator **generator_1** is used to generate the sequence **random_array_1**, which is a sequence of 1,000,000 random integers in the range [0, 1,000[.

Meanwhile, the **generator_2** generator is used to generate the sequence **random_array_2**, which is also a sequence of 1,000,000 random integers in the range [0, 1,000[.

Finally, the sequences **random_array_1** and **random_array_2**, generated from different generators, but with the same seed, are printed.

**Conclusion:**

Note that the resulting sequences are the same. In this way, the use of the same seed allows the reproduction of the results.

```python
seed_1 = 1

generator_1 = Generator(PCG64(seed_1))
random_array_1 = generator_1.integers(low=0, high=1_000, size=1_000_000)
print("random_array_1: ", random_array_1)

generator_2 = Generator(PCG64(seed_1))
random_array_2 = generator_2.integers(low=0, high=1_000, size=1_000_000)
print("random_array_2: ", random_array_2)
```

Thus, the graphs in Figure 1 show the histograms of a sample of a **"discrete uniform"** distribution with period $2^{128}$ and seed **seed_1**.

**Conclusion:**

Note that the two histograms are identical, since they were generated from two sequences generated with the same seed.

```python
fig, axs = plt.subplots(2, 1, figsize=(7, 10)) #rows = 2, cols = 1

axs[0].set_title("Graph 1 - Histogram of 'random_array_1'.")
axs[0].set_xlabel('Sequence value')
axs[0].set_ylabel('Frequency')
axs[0].hist(random_array_1, bins = 1000, color='tab:blue', alpha=0.7)

axs[1].set_title("Graph 2 - Histogram of 'random_array_2'.")
axs[1].set_xlabel('Sequence value')
axs[1].set_ylabel('Frequency')
axs[1].hist(random_array_2, bins = 1000, color='tab:blue', alpha=0.7)

fig.suptitle('Figure 1 - Sequences with the same seed.')

fig.tight_layout()
fig.subplots_adjust(top=0.92)

plt.show()
```

# 2.Descriptive statistic

*   For a 5-numbers sequence: min, average, mode, median, max, range, sum, variance, standard deviation;
*   For a 100-numbers sequence: min, average, mode, median, max, range, sum, variance, standard deviation;
*   Distribution(controlling bins), and percentiles;
*   Average as a weighted sum.

 **Description of Additional Libraries:**

The **numpy** library [8] was mainly used to calculate basic functions of descriptive statistics.

The **scipy.stats** library [9] was used specifically for the mode calculation.

The **collections** library [10] was used to create a dictionary data structure.

```python
import numpy as np
from scipy.stats import mode
from collections import Counter
```

**Description of code:**

Below are defined the basic functions of descriptive statistics that will be used throughout this work: min, mean, mode, median, max, range, sum, variance (var), standard deviation (std).

Furthermore, a way to handle the average as a weighted sum (calc_average_as_weighted_sum) was defined, which consists of selecting the distinct elements in an array, and using their frequency as the weight of the weighted average.

Finally, a function was defined to calculate the percentiles (calc_percentile), where we chose to use the parameter interpolation="midpoint", as explained through comments in the source code.

```python
def calc_min(array):
  return np.min(array)

def calc_max(array):
  return np.max(array)

def calc_range(array):
  return np.max(array) - np.min(array)

def calc_sum(array):
  return np.sum(array)

def calc_var(array):
  return np.var(array)

def calc_std(array):
  return np.std(array)  

def calc_median(array):
  return np.median(array)

def calc_mode(array):
  return mode(array)

def calc_mean(array):
  return np.mean(array)

def calc_average_as_weighted_sum(array):
  distinct_items_array =[]
  weights = []
  #Returns a dictionary data structure where the 'keys' are the array elements and the 'values' are their respective frequencies
  value_weights_array = Counter(array)
  for (key, value) in value_weights_array.items():
    distinct_items_array.append(key)
    weights.append(value)
  return np.average(distinct_items_array, weights=weights)

def calc_percentile(input_array, percentile_array):
  #Options for the 'interpolation' parameter, used when the desired percentile is between two data points i < j: 
  #{'linear', 'lower', 'higher', 'midpoint', 'nearest'}
  #Chosen: 'midpoint': (i + j) / 2. 
  return np.percentile(input_array, percentile_array, interpolation="midpoint")
```

**Description of code:**

The code below generates a sequence of only 5 integers (random_array_6), in the range of [1,10]. After that, the descriptive statistics functions for the generated sequence are calculated and printed.

```python
seed_6 = 6
generator_6 = Generator(PCG64(seed_6))
random_array_6 = generator_6.integers(low=1, high=10, size=5, endpoint=True)

print("Sequence of 5 numbers: ", random_array_6)
print("Min: ", calc_min(random_array_6))
print("Max: ", calc_max(random_array_6))
print("Range: ", calc_range(random_array_6))
print("Sum: ", calc_sum(random_array_6))
print("Variance: ", calc_var(random_array_6))
print("Standard Deviation: ", calc_std(random_array_6))
print("Median: ", calc_median(random_array_6))
print("Mode: ", calc_mode(random_array_6))
print("Average: ", calc_mean(random_array_6))
print("Average as a weighted sum: ", calc_average_as_weighted_sum (random_array_6))
print("Percentiles [0,5,25,50,75,80,95,100]: ", calc_percentile(random_array_6, [0,5,25,50,75,80,95,100]))
```

## 3.1. Time Series 

**Description of code:**

First, the code below generates a sequence of 100 dates starting on December 1, 2021 and incrementing by 1 day.

After that, it creates another sequence of 100 random integer numbers varying in the range of [1,5] ('random_array_8').

As an example, the sequence 'random_array_7' was used to represent the hydrological alerts issued daily, and the sequence 'random_array_8' to represent the geological alerts.

Finally, two time series were created: timeseries_hydro and timeseries_geo. 

```python
#Define the number of dates
n=100
# Set the start date
date = datetime.datetime(2021,12,1)
#Create the date array and add the start date
dates = []
dates.append(date)
for i in range(n-1): 
    date += datetime.timedelta(days=1)
    dates.append(date) 

#Create another sequence, the Geological alerts sequence
seed_8 = 8
generator_8 = Generator(PCG64(seed_8))
random_array_8 = generator_8.integers(low=1, high=5, size=100, endpoint=True)

timeseries_hydro = {'date': dates,
                'alert': random_array_7}

timeseries_geo = {'date': dates,
                'alert': random_array_8}
```

```python
fig, axs = plt.subplots(figsize=(18,6))

axs.plot(timeseries_hydro['date'], timeseries_hydro['alert'], color='tab:blue', label='Hydrological', linewidth=3)
axs.plot(timeseries_geo['date'], timeseries_geo['alert'], color='tab:brown', label='Geological', linewidth=3)

plt.xlabel('Date')
plt.ylabel('Number of Alerts')

plt.legend()

fig.suptitle("Figure 9 - Time Series of Sent Alerts.") 

fig.tight_layout()
fig.subplots_adjust(top=0.94)

plt.show() 
```

```python
#Create sequential subsets of length 'size'
def calc_slices(array, size):
  slices_array = []
  i=0
  while(i < len(array)):
    slices_array.append(array[i:i+size])
    i+=size
  return slices_array
```

# 4.Geometry

The purpose of this section is to try to graphically visualize some metrics of descriptive statistics.

Metrics Explanation:

*   Min is the smallest value of the data;
*   Max is the largest value of the data;
*   Range is the difference between Max and Min;
*   The sample mean is the average and is computed as the sum of all the observed outcomes from the sample divided by the total number of events;
*   The median is the middle score.  If we have an even number of events we take the average of the two middles.  The median is better for describing the typical value;
*   Mode is the number which occur most often in the data;
*   Variance is the numerical values that describe the variability of the observations from its mean and denoted by sigma-squared (σ2). Variance measure how far individuals in the group are spread out, in the set of data from the mean;
*   Standard Deviation is a measure of dispersion of observation within data relative to their mean. It is square root of the variance and denoted by sigma (σ).

**Description of code:**

# 6.Moving Average

**Description of Additional Libraries:**

The **pandas** library [12] was used to create a time series and calculate different moving averages.

```python
from pandas import Series
```

**Description of code:**

First the numpy array is converted into a time series object and then the rolling() function is used to perform the calculation in the rolling window and calculate the moving average using the mean() function.

Three windows are calculated, with the following sizes: 5(called 'short_rolling' here), 10('medium_rolling'), 15('long_rolling').

Note that if the time window range is n, there are n-1 NaN values at the beginning because the moving average could not be calculated for them.

```python
data = random_array_7
series = Series(data)

# Calculating the short-window simple moving average
short_rolling = series.rolling(window=5).mean()
# Calculating the medium-window simple moving average
medium_rolling = series.rolling(window=10).mean()
# Calculating the long-window simple moving average
long_rolling = series.rolling(window=15).mean()

print("short_rolling: ", short_rolling)
print("medium_rolling: ", medium_rolling)
print("long_rolling", long_rolling)
```

**Description of code:**

Figure 17 shows the time series of hydrological alerts from Figure 9, with three moving averages and different types of windows.

```python
fig, axs = plt.subplots(figsize=(18,6)) 

axs.plot(timeseries_hydro['date'], timeseries_hydro['alert'], color='tab:blue', label='Hydrological Alerts', linewidth=2)
axs.plot(timeseries_hydro['date'], short_rolling, color='tab:red', label='short_rolling: 5-days SMA', linewidth=2)
axs.plot(timeseries_hydro['date'], medium_rolling, color='tab:orange', label='medium_rolling: 10-days SMA', linewidth=2)
axs.plot(timeseries_hydro['date'], long_rolling, color='tab:purple', label='long_rolling: 15-days SMA', linewidth=2)

plt.xlabel('Data')
plt.ylabel('Number of Alerts')

plt.legend(loc='best')

fig.suptitle("Figure 17 - Different Moving Averages for the Time Series of Sent Hydrological Alerts.")  

fig.tight_layout()
fig.subplots_adjust(top=0.94)

plt.show() 
```

**Conclusions:**

* If the window size is 1, the moving average becomes the time series itself;
* The larger the window size, the greater the smoothing obtained on the moving average.
